```python
# import tkinter as tk
import random
# original code and idea I modified from https://stackoverflow.com/questions/7214054/tkinter-making-a-second-canvas-display-the-contents-of-another
class LabelGui(tk.Tk):
    def __init__(self, *args, **kwargs):
        
        tk.Tk.__init__(self, *args, **kwargs)
        self.previous_x = self.previous_y = 0
        self.x = self.y = 0
        self.points_recorded = []
        # Declare Peered Canvas object, from class defined below
        # Peered Canvas methods mirror widget functionality across multiple canvas images
        self.canvas1 = PeeredCanvas(self, width=400, height=400, bg="black", border=1, relief="sunken",cursor="cross")
        self.canvas2 = PeeredCanvas(self, width=400, height=400, bg="black",border=1, relief="sunken",cursor="cross")
        self.canvas3 = PeeredCanvas(self, width=400, height=400, bg="black",border=1, relief="sunken",cursor="cross")
        self.canvas1.add_peer(self.canvas2)
        self.canvas1.add_peer(self.canvas3)
        toolbar = tk.Frame(self)
        clear_button = tk.Button(self, text="Clear Points", command=self.clear_all_points)
        clear_button.pack(in_=toolbar, side="left")
        
        # add buttons here for 'adding points', 'removing points', 'saving points'
        toolbar.pack(side="top", fill="x")
        self.canvas1.pack(side="left", fill="both", expand=True)
        self.canvas2.pack(side="left", fill="both", expand=True)
        self.canvas3.pack(side="left", fill="both", expand=True)
        self.canvas1.bind("<q>", self.export_work)
        self.canvas1.bind("<Button-1>", self.draw_from_where_you_are)
    ### Monkey Patch for tkinter_circle routine
    
    def create_circle(self, x, y, r, **kwargs):
        return self.canvas1.create_oval(x-r, y-r, x+r, y+r, **kwargs)

    def export_work(self):
        
        self.canvas1.create_line(self.previous_x, self.previous_y, 
                                self.x, self.y,fill="yellow")
    def tell_me_where_you_are(self, event):
        self.previous_x = event.x
        self.previous_y = event.y
        
    def draw_from_where_you_are(self, event):
        if self.points_recorded:
            self.points_recorded.pop()
            self.points_recorded.pop()

        self.x = event.x
        self.y = event.y
        self.canvas1.create_line(self.x+2, self.y+2, 
                                self.x, self.y,fill="yellow")
        self.points_recorded.append(self.previous_x)
        self.points_recorded.append(self.previous_y)
        self.points_recorded.append(self.x)     
        self.points_recorded.append(self.x)        
        self.previous_x = self.x
        self.previous_y = self.y
        
                                    
    #        self.points_recorded.append(self.x)
     #       self.points_recorded.append(self.y)
        
    
    def circle_conversion(self):
        x0=self.points_recorded[0]
        x1=self.points_recorded[2]
        y0=self.points_recorded[1]
        y1=self.points_recorded[3]
        mid=((x1-x0)/2,(y1-y0)/2)
        r=np.sqrt((x0+mid[0])**2+(y0+mid[1])**2)
        return r
        # functions added from previous generations 
    def onQuit(self):
        self.window.destroy()
      
    def clear_all_points(self):
        self.canvas1.delete("all")
        
        
    
# this is a peer class of canvas that allows us to modify objects on other canvases by interacting with a seperate canvas
# this technique allows us to label one image and update images on multiple others

class PeeredCanvas(tk.Canvas):
    '''A class that duplicates all objects on one or more peer canvases'''
    def __init__(self, *args, **kwargs):
        self.peers = []
        tk.Canvas.__init__(self, *args, **kwargs)

    def add_peer(self, peer):
        if self.peers is None:
            self.peers = []
        self.peers.append(peer)
    # second round of altered code here
    def clear_all_points(self):
        tk.Canvas.delete(self,*args,**kwargs)
        for peer in self.peers:
            peer.delete(*args,**kwargs)
            
    def bind(self,*args,**kwargs):
        tk.Canvas.bind(self,*args,**kwargs)
        for peer in self.peers:
            peer.bind(*args,**kwargs)
            
    def create_line(self,*args,**kwargs):
        tk.Canvas.create_line(self,*args,**kwargs)
        for peer in self.peers:
            peer.create_line(*args,**kwargs)
    # altered code here
    def make_draggable(self,*args,**kwargs):
        widget.bind(self,*args,**kwargs)
        #widget.bind(self,*args,**kwargs)
        for peer in self.peers:
            peer.widget.bind(self,*args,**kwargs)
            
    def on_drag_start(self,*args,**kwargs):
        event.widget._drag_start_x = self.event.x
        event.widget._drag_start_y = self.event.y
        for peer in self.peers:
            peer.event.widget._drag_start_x = self.event.x
            peer.event.widget._drag_start_y = self.event.y

    def on_drag_motion(self,*args,**kwargs):
        widget = event.widget
        x = widget.winfo_x() - widget._drag_start_x + self.event.x
        y = widget.winfo_y() - widget._drag_start_y + self.event.y
        widget.place(x=x, y=y)
        for peer in self.peers:
            peer.widget.place(x=x,y=y)
        
    # end new code  
    def move(self, *args, **kwargs):
        tk.Canvas.move(self, *args, **kwargs)
        for peer in self.peers:
            peer.move(*args, **kwargs)

    def itemconfigure(self, *args, **kwargs):
        tk.Canvas.itemconfigure(self, *args, **kwargs)
        for peer in self.peers:
            peer.itemconfigure(*args, **kwargs)

    def delete(self, *args, **kwargs):
        tk.Canvas.delete(self, *args)
        for peer in self.peers:
            peer.delete(*args)

    def create_oval(self, *args, **kwargs):
        tk.Canvas.create_oval(self, *args, **kwargs)
        for peer in self.peers:
            peer.create_oval(*args, **kwargs)


app = LabelGui()
app.mainloop()
```
