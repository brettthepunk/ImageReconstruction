```python
# Code here is workspace for Labeling Gui
# Code is not intended for use in lab, is only for dev work

#### LabelMaster BETA ###

# Import libraries

import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
import random
import os
import math
from scipy.spatial import distance
from tkinter import filedialog
import numpy as np
import time

#from tkFileDialog import *
# global variables
dataDir = os.getcwd()    
filetype='.png'
armsize=40
wristsize=20
digitsize=10
pixwidth=1280/2
pixheight=1080/2
digitnum=2
arm=0
wrist=0
digit=0
scorer="Brett"
bodyparts="digit","arm","wrist"
class LabelGui(tk.Tk):
    def __init__(self, *args, **kwargs):
        global start # start variable for angular motion
        tk.Tk.__init__(self, *args, **kwargs)
    
        # initialize variables
        self.previous_x = self.previous_y = 0
        self.x = self.y = 0
        self.points_recorded = []
        self.grid()
        self.bttn_clicks = 0 
        self.arm=0
        self.wrist=0
        self.digit=0
       

        # declaring tk StringVars
        oc = tk.StringVar(self)
        oc.set("Select")
        folder_path = tk.StringVar()
        self.dataDir = tk.StringVar()
        self.dataDir.set(dataDir) 
        
        # Declare Peered Canvas object, from class defined below
        # Peered Canvas methods mirror widget functionality across multiple canvas images
        self.canvas1 = PeeredCanvas(self, width=pixwidth, height=pixheight, bg="black", border=1, relief="sunken",cursor="cross")
        self.canvas2 = PeeredCanvas(self, width=pixwidth, height=pixheight, bg="black",border=1, relief="sunken",cursor="cross")
        self.canvas3 = PeeredCanvas(self, width=pixwidth, height=pixheight, bg="black",border=1, relief="sunken",cursor="cross")
        self.canvas1.add_peer(self.canvas2)
        self.canvas1.add_peer(self.canvas3)
        toolbar = tk.Frame(self)
        # BUTTONS
        clear_button = tk.Button(self, text="Clear Points", command=self.clear_all_points)
        clear_button.pack(in_=toolbar, side="left")
        o = tk.OptionMenu(self, oc, "arm","wrist","digit", command=self.line_select)
        o.pack(in_=toolbar,side="right")
        btn_label=tk.Button(self,text="Label Points",width=30,command=self.label_points)
        btn_label.pack(in_=toolbar,side="right")
        # note: Browse_btn must be debugged 
        browse_btn=tk.Button(text="Browse Files",width=35, command=self.ddBrowseCallback)
        browse_btn.pack(in_=toolbar,side="left")
        btn_next=tk.Button(text="Next Image",width=30,command=self.image_return)
        btn_next.pack(in_=toolbar,side="right")
        btn_transform=tk.Button(self,text='Translation Movement', width=30,command=self.translation_onclick)
        btn_transform.pack(in_=toolbar,side="left")
        btn_populate=tk.Button(self,text="Initialize Images",width=40,command=self.populate_canvas)
        btn_populate.pack(in_=toolbar,side="right")
        btn_rotation=tk.Button(self,text="Rotation Movement",width=30,command=self.rotation_onclick)
        btn_rotation.pack(in_=toolbar,side="left")
        btn_savepoints=tk.Button(self,text="Save Points",width=30,command=self.savepoints)
        btn_savepoints.pack(in_=toolbar,side="left")
        # add buttons here for 'adding points', 'removing points', 
        toolbar.pack(side="top", fill="x")
        self.canvas1.pack(side="left", fill="both", expand=True)
        self.canvas2.pack(side="left", fill="both", expand=True)
        self.canvas3.pack(side="left", fill="both", expand=True)
        self.canvas1.bind("<ButtonPress-1>", self.onPressToMove)
        self.canvas1.bind( "<B1-Motion>", self.onMovement )
        #self.canvas1.bind("<Button-2>",self.rotation_onclick) # possible in more advanced versions...
        
    def populate_canvas(self):
        # populate canvas with images for testing/showing off functionality
        self.my_image = tk.PhotoImage(file='rat.png')
        self.canvas1.image = self.my_image
        self.canvas1_image=self.canvas1.create_image(0,0,image=self.canvas1.image,anchor='nw',tag="canvas")
        
        self.canvas2_image=self.canvas2.create_image(40,40,image=self.my_image,anchor='nw',tag="canvas1")
        self.canvas3_image=self.canvas3.create_image(-20,-20,image=self.my_image,anchor='nw',tag="canvas2")
    ###### Code Below is Exclusively for Drag Functionality ######
    def translation_onclick(self):  
        # Bind relevant functions
        self.canvas1.bind("<ButtonPress-1>", self.onPressToMove)
        self.canvas1.bind( "<ButtonRelease-1>", self.on_drop)
        self.canvas1.bind( "<B1-Motion>", self.onMovement )
    
    
    def onPressToMove(self, event): #get initial location of object to be moved
        winX = event.x 
        winY = event.y 
        self.widget_item= self.canvas1.find_closest(event.x, event.y)# find object ID
        self.tags = self.canvas1.itemcget(self.widget_item, "tags") # see if this item contains any of our tags
        if "canvas" in self.tags:
            messagebox.showinfo("Click on the segment")
            pass
        else:
            self.canvas1.tag_raise(self.widget_item)# bring this item to front of canvas
            self.previous_x = winX
            self.previous_y = winY
    def onMovement(self, event):
        winX = event.x
        winY = event.y 
        self.newX = winX - self.previous_x
        self.newY = winY - self.previous_y
        
        self.canvas1.move(self.widget_item, self.newX/100, self.newY/100)
        time.sleep(0.01) # delay between movements (?)
    def on_drop(self,event):
        winX = event.x 
        winY = event.y 
        self.newX = winX - self.previous_x
        self.newY = winY - self.previous_y
        if self.widget_item is not None:
            self.canvas1.move(self.widget_item,self.newX,self.newY)
        
#### Code Below is Exclusively for Image Selection/Browsing #####

################# WIP ##############
    def image_select(self):
        
        self.file_choice=random.choice(self.filecontains) # selects random file
        self.prefix_stamp=self.file_choice[14:26]
        #[x for x in if prefix_stamp not in x] # deletes all files containing portion
    def assign_images_to_windows(self):
        for datafile in os.listdir(str(self.dataDir.get())):
             if self.prefix_stamp in datafile and datafile.endswith(filetype):
                    if "camera: 0" in datafile:
                        self.img_name=datafile
                        self.my_image1 = tk.PhotoImage(dataDir+datafile)
                        self.canvas1.create_image(pixwidth,pixheight,image=self.my_image1,anchor='nw') # create image here for canvas1
                    if "camera: 1" in datafile:
                        self.img_name=datafile
                        self.my_image2=tk.PhotoImage(dataDir+datafile)
                        self.canvas2.create_image(pixwidth,pixheight,image=self.my_image2,anchor='nw') # create image here for canvas2
                    if "camera: 2" in datafile:
                        self.img_name=datafile
                        self.my_image3=tk.PhotoImage(dataDir+datafile)
                        self.canvas3.create_image(pixwidth,pixheight,image=self.my_image3,anchor='nw') # create image here for canvas3
                    else:
                        messagebox.showinfo("Warning", "you are attempting to populate with a empty directory. Select Another"
                                           )
    def image_return(self):
        os.chdir(dataDir)
        self.filecontains=os.listdir()
        self.image_select()
        self.assign_images_to_windows()
        # take first n digits and 
        
        
    def ddBrowseCallback(self):
        
        self.dataDir.set(filedialog.askdirectory())
        
### Code Below is Exclusively for Rotational Transform of Objects ###
    def rotation_onclick(self):
            # Bindings for rotation
        self.canvas1.bind("<Button-1>", self.press)
        self.canvas1.bind("<B1-Motion>", self.motion)
        self.canvas1.unbind( "<ButtonRelease-1>")

    def getangle(self,event):
        self.dx = self.canvas1.canvasx(event.x) - self.center[0] # get dx
        self.dy = self.canvas1.canvasy(event.y) - self.center[1] # get dy
        try:
            return complex(self.dx, self.dy) / abs(complex(self.dx, self.dy)) # get angle 
        except ZeroDivisionError:
            return 0.0 # cannot determine angle

    def press(self,event):
        # calculate angle at start point
        self.polygon_item= self.canvas1.find_closest(event.x, event.y) # Find i
        self.tags = self.canvas1.itemcget(self.polygon_item, "tags") # see if this item contains any of our tags
        if "canvas" in self.tags:
            messagebox.showinfo("Click on the segment")
            pass
        else:
            
            self.canvas1.tag_raise(self.polygon_item) # put this item on top 
            initial_center = self.canvas1.coords(self.polygon_item) # Find item coords
            # take absolute distance between event and endpoints
            end_1=np.sqrt((initial_center[0]-event.x)**2+(initial_center[1]-event.y)**2)
            end_2=np.sqrt((initial_center[2]-event.x)**2+(initial_center[3]-event.y)**2)
            # set center to closest of endpoints for angular movement
            if end_1 > end_2:
                self.center=[initial_center[0],initial_center[1]]
            else:
                self.center=[initial_center[2],initial_center[3]]
            self.start = self.getangle(event)
        
        
    def motion(self,event):
        # calculate current angle relative to initial angle
        self.angle = self.getangle(event) / self.start # obtain current angle
        self.offset = complex(self.center[0], self.center[1]) # find center offset
        newxy = [] 
        xy=self.canvas1.coords(self.polygon_item) # get current coordinates of object
        self.v1=self.angle * (complex(xy[0],xy[1])-self.offset)+self.offset # calculate vector of first endpoint
        newxy.append(self.v1.real)
        newxy.append(self.v1.imag)
        self.v2=self.angle * (complex(xy[2],xy[3])-self.offset)+self.offset  #calculate vector of second endpoint
        newxy.append(self.v2.real)
        newxy.append(self.v2.imag)
        self.canvas1.coords(self.polygon_item, *newxy)  #update with new coordinates
        # need to update the other two with relative rotational transformations
        # hard-code for now, this needs to call into a function
        self.canvas2.coords(self.polygon_item,*newxy)
        self.canvas3.coords(self.polygon_item,*newxy)
        time.sleep(0.01) # this code is to attempt to slow how fast the mouse moves, try something different tbh
### Code Below is for Saving and Appending Points Between Frames/Sessions ####
    
    
    def frame_savepoints(self):
        self.scorer=scorer
        self.bodyparts=bodyparts
        # possible save headers here to store as a key..
        # pull up and open .h5 file storing file for experiment run
        for datafile in os.listdir(str(self.dataDir.get())):
            if self.prefix_stamp in datafile and datafile.endswith('.h5'):
                    self.dataframe=pd.read_hdf(datafile)
            else:# create from scratch
                df = pd.DataFrame(rows=['Scorer','bodyparts','coords'])
                
                
                
                #self.dataFrame.loc[self.img_name,self.scorer, enumerate over bodyparts for coordinates]
        
    def savepoints(self):
        pass
    
### Code Below is for Segment Selection Button Menu Exclusively ###
    def line_select(self,input):
        # takes input from listbox as string, creates line of whatever size is specified 
        
        if input == "arm" and self.arm ==0:
            self.arm_segment=self.canvas1.create_line(180+armsize, 180, 180, 180, fill='red', width=2)
            self.canvas1.tag_raise(self.arm_segment)
            self.arm=1
        if input == "arm" and self.arm ==1:
            messagebox.showinfo("Warning", "Why are you trying to put multiple arm segments?")
        if input == "wrist"and self.wrist ==0:
            self.wrist_segment=self.canvas1.create_line(100+wristsize, 100, 100,100, fill='red', width=2)
            self.canvas1.tag_raise(self.wrist_segment)
            self.wrist=1
        if input == "wrist" and self.wrist ==1:
            messagebox.showinfo("Warning", "Why are you trying to put multiple wrist segments?")
        if input=="digit" and self.digit ==0:
            self.digit_segment=self.canvas1.create_line(10+digitsize, 10, 10, 10, fill='red', width=2)
            self.canvas1.tag_raise(self.digit_segment)
            self.digit+=1
        if input == "digit" and self.digit > digitnum:
            messagebox.showinfo("Warning", "Why are you trying to put more digit segments?")
  ### Code Below is for Point Labeling w/ Line's Exclusively ####
    
    def update_count(self):
        if self.bttn_clicks <2:
            self.bttn_clicks += 1
        else:
            self.bttn_clicks=1
        
    def tell_me_where_you_are(self, event):
        self.previous_x = event.x
        self.previous_y = event.y
    def label_points(self):
        self.canvas1.bind("<Button-1>", self.draw_from_where_you_are)
        self.canvas1.bind( "<B1-Motion>", self.tell_me_where_you_are )
        self.canvas1.unbind( "<ButtonRelease-1>")
    def draw_from_where_you_are(self, event):
        self.update_count()
        self.x = event.x
        self.y = event.y
        if self.bttn_clicks==2:
            self.canvas1.create_line(self.previous_x, self.previous_y, 
                                self.x, self.y,fill="yellow")
            self.canvas1.create_oval(self.previous_x,self.previous_y,self.x,self.y)
        
       
        
        if self.points_recorded:
            self.points_recorded.pop()
            self.points_recorded.pop()
        
        
        self.canvas1.create_line(self.x+2, self.y+2, 
                                self.x, self.y,fill="yellow")
        self.points_recorded.append(self.previous_x)
        self.points_recorded.append(self.previous_y)
        self.points_recorded.append(self.x)     
        self.points_recorded.append(self.y)        
        self.previous_x = self.x
        self.previous_y = self.y
        
        
     
    def delete_point(self,event):
        event_point= self.canvas1.find_closest(event.x, event.y)# Find event
        initial_center = self.canvas1.coords(self.polygon_item) # get coordinates of event.
        # calculate endpoints here
        end_1=np.sqrt((initial_center[0]-event.x)**2+(initial_center[1]-event.y)**2) 
        end_2=np.sqrt((initial_center[2]-event.x)**2+(initial_center[3]-event.y)**2)
        # delete event points inside recorded points if one endpoint is closer than the other
        if end_1 > end_2:
            if initial_center[0] in self.points_recorded: self.points_recorded.remove(initial_center[0])
            if initial_center[1] in self.points_recorded: self.points_recorded.remove(initial_center[1])
        else:
            if initial_center[2] in self.points_recorded: self.points_recorded.remove(initial_center[2])
            if initial_center[3] in self.points_recorded: self.points_recorded.remove(initial_center[3])
        self.canvas1.delete(event_point) # delete the event
### Code here is utilities/smaller functions ####     
    def clear_all_points(self):
        self.canvas1.delete("all") # delete all canvas objects
    def onQuit(self):
        self.window.destroy()  
        
    
# this is a peer class of canvas that allows us to modify objects on other canvases by interacting with a seperate canvas
# this technique allows us to label one image and update images on multiple others



class PeeredCanvas(tk.Canvas):
    '''A class that duplicates all objects on one or more peer canvases'''
    def __init__(self, *args, **kwargs):
        self.peers = []
        tk.Canvas.__init__(self, *args, **kwargs)

    def add_peer(self, peer):
        if self.peers is None:
            self.peers = []
        self.peers.append(peer)
    # second round of altered code here
    # This round is for binding/creation
    def clear_all_points(self):
        tk.Canvas.delete(self,*args,**kwargs)
        for peer in self.peers:
            peer.delete(*args,**kwargs)
    def bind(self,*args,**kwargs):
        tk.Canvas.bind(self,*args,**kwargs)
        for peer in self.peers:
            peer.bind(*args,**kwargs)

    
    def create_line(self,*args,**kwargs):
        tk.Canvas.create_line(self,*args,**kwargs)
        for peer in self.peers:
            peer.create_line(*args,**kwargs)
    # altered code here
    # this code was intended for drag functionality, is not needed at the moment
    def make_draggable(self,*args,**kwargs):
        tk.widget.bind(self,*args,**kwargs)
        #widget.bind(self,*args,**kwargs)
        for peer in self.peers:
            peer.widget.bind(self,*args,**kwargs)
            
    def on_drag_start(self,*args,**kwargs):
        tk.event.widget._drag_start_x = self.event.x
        tk.event.widget._drag_start_y = self.event.y
        for peer in self.peers:
            peer.event.widget._drag_start_x = self.event.x
            peer.event.widget._drag_start_y = self.event.y

    def on_drag_motion(self,*args,**kwargs):
        widget = tk.event.widget
        x = widget.winfo_x() - widget._drag_start_x + self.event.x
        y = widget.winfo_y() - widget._drag_start_y + self.event.y
        widget.place(x=x, y=y)
        for peer in self.peers:
            peer.widget.place(self, *args,**kwargs)
    
    # More motivated code from example 
    def move(self, *args, **kwargs):
        tk.Canvas.move(self, *args, **kwargs)
        for peer in self.peers:
            peer.move(*args, **kwargs)

    def itemconfigure(self, *args, **kwargs):
        tk.Canvas.itemconfigure(self, *args, **kwargs)
        for peer in self.peers:
            peer.itemconfigure(*args, **kwargs)
            
    def delete(self, *args, **kwargs):
        tk.Canvas.delete(self, *args)
        for peer in self.peers:
            peer.delete(*args)
    def create_oval(self,*args,**kwargs):
        tk.Canvas.create_oval(self, *args, **kwargs)
        for peer in self.peers:
            peer.create_oval(*args, **kwargs)


app = LabelGui()
app.mainloop()
```
