```python
import h5py
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy

filename = 'trial.h5'
```

```python
df=pd.read_hdf(filename)
df.columns=df.columns.droplevel(0)
```

```python
df.keys()
```

```python
Wrist2=df.loc[:, df.columns.get_level_values(0).isin(['Wristr2'])]   #
xwr=Wrist2['Wristr2','x'].copy()
ywr=Wrist2['Wristr2','y'].copy()
probwr=Wrist2['Wristr2','likelihood'].copy()
#xwr=xwr-600
    

comp_time=np.linspace(0,len(ywr),len(ywr))
fig, axs = plt.subplots(nrows=4,figsize=(17,17))
plt.suptitle('Positions and probabilities for Right Wrist captured by Camera 2')
sns.scatterplot(comp_time,xwr, ax=axs[0],label='X position',color='red')
plt.xlabel('Frames')
plt.ylabel('Position (X)')
plt.title('X Position over frames')
sns.scatterplot(comp_time,ywr, ax=axs[1],label='Y Position',color='orange')
plt.xlabel('Frames')
plt.ylabel('Position (Y)')
plt.title('Y position over frames')
sns.scatterplot(comp_time,probwr,ax=axs[2],label='Probabilities',color='green')
plt.xlabel('Frames')
plt.ylabel('Probability (1-P)')
plt.title('1-p probability value for chosen points')
plt.xlabel('')
sns.scatterplot(xwr,ywr, ax=axs[3],label='XY position',color='black')
plt.xlabel('Position(Y)')
plt.ylabel('Position (X)')
plt.title('XY positions over frames')

plt.savefig('RightWrist2_total.png')
```
